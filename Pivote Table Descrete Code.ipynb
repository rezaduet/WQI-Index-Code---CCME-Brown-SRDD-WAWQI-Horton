#Merge Multiple CSV files into in From a Google Drive folder

```python
from google.colab import drive
import pandas as pd
import glob
import os

# Mount your Google Drive
drive.mount('/content/drive')

# Set the directory where your CSV files are stored in your Google Drive
directory = '/content/drive/MyDrive/rifat/Ballyteigue-Bannow'



# Print the contents of the directory to check if the directory path is correct
print(os.listdir(directory))

# Get a list of all CSV files in the directory
csv_files = glob.glob(directory + '/*.csv')


# Print the list of CSV files found in the directory
print(csv_files)

# Initialize an empty list to hold the DataFrames
dfs = []

# Loop over each CSV file and read it into a DataFrame
for filename in csv_files:
    df = pd.read_csv(filename)
    dfs.append(df)

# Print the list of DataFrames to check if the DataFrames are being read in correctly
print(dfs)

# Concatenate all of the DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
# Save the merged DataFrame as a CSV file
merged_df.to_csv('/content/drive/MyDrive/Ballyteigue-Bannow.csv', index=False)
```
